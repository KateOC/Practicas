```python
def descargar_comentarios_video(yt,videoId):
    request = youtube.commentThreads().list(
        part="snippet,replies",videoId= videoId)
    response = request.execute()
    respuestas = []
    respuestas.append(response)
    next_page_token = response.get('nextPageToken')
    mas_pag = not next_page_token is None
    while mas_pag:
      request = youtube.commentThreads().list(
                  part="snippet,replies",
                  videoId= videoId,
                  pageToken = next_page_token)
      response = request.execute()
      respuestas.append(response)
      next_page_token = response.get('nextPageToken')
      mas_pag = not next_page_token is None
    return respuestas
```

```python
def descargar_canal_ids(yt, canalId):
    request = yt.search().list(
        part="id",
        channelId=canalId,
        maxResults=100)
    response = request.execute()
    lista_ids = []
    items = response['items']
    for item in items:
      try:
        id = item['id'] ['videoId']
        lista_ids.append(id)
      except:
        continue
    next_page_token = response.get('nextPageToken')
    mas_pag = not next_page_token is None
    while mas_pag:
      request = yt.search().list(
                    part="id",
                    channelId= canalId,
                    maxResults = 50,
                    pageToken = next_page_token)
      response = request.execute()
      items = response['items']
      for item in items:
        try:
          id = item['id'] ['videoId']
          lista_ids.append(id)
        except:
          continue
      next_page_token = response.get('nextPageToken')
      mas_pag = not next_page_token is None
    return lista_ids
```

```python
def ordenar_info_video(yt,videoId):
  info_video = descargar_comentarios_video(yt,videoId)
  dicc = {"id" : videoId}
  comentarios = []
  for comentario in info_video:
    its = comentario['items']
    for item in its:
      dicc_comentario = {}
      snp = item['snippet']['topLevelComment']['snippet']['textOriginal']
      dicc_comentario["comentario"] = snp
      num_respuestas = item['snippet']['totalReplyCount']
      respuestas = []
      if num_respuestas > 0:
        respuestas_orig = item['replies']['comments']
        for respuesta in respuestas_orig:
          its_respuesta = respuesta['snippet']['textOriginal']
          respuestas.append(its_respuesta)
      dicc_comentario["respuestas"] = respuestas
      comentarios.append(dicc_comentario)
  dicc["comentarios"] = comentarios
  return dicc
```

```python
def ordenar_info_canal(yt, canalId, maxVids):
  videos_id = descargar_canal_ids(yt, canalId)
  info = {"canal" : canalId}
  i = 0
  videos_info = []
  for video_id in videos_id:
    info_video = ordenar_info_video(yt, video_id)
    videos_info.append(info_video)
    if i >= maxVids:
      break
    i = i + 1
  info["videos"] = videos_info
  return info
```

```python
#Prueba de almacenar comentarios en ficheros de texto de la lista de reproducción "Traditionalism" de Mrs. Midwest
import googleapiclient.discovery

# Clave de la API de YouTube
API_KEY = "AIzaSyBS7HnwtP1DeUDjO_bMYKH-qcJZYJuJ2dc"

# ID de la lista de reproducción de YouTube
PLAYLIST_ID = "PLss0S_sx8-ukieVx7ANEAJ7DtP1tah4Ms"

# Crea un objeto de servicio de la API de YouTube
youtube = googleapiclient.discovery.build('youtube', 'v3', developerKey=API_KEY)

# Función para obtener los IDs de todos los videos en la lista de reproducción
def obtener_ids_videos():
    videos_ids = []
    next_page_token = None
    while True:
        request = youtube.playlistItems().list(
            part="contentDetails",
            playlistId=PLAYLIST_ID,
            maxResults=50,  # Máximo de resultados por página
            pageToken=next_page_token
        )
        response = request.execute()
        for item in response["items"]:
            videos_ids.append(item["contentDetails"]["videoId"])
        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break
    return videos_ids

# Función para obtener los comentarios de un video
def obtener_comentarios(video_id):
    comentarios = []
    next_page_token = None
    while True:
        request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=100,  # Máximo de comentarios por página
            pageToken=next_page_token
        )
        response = request.execute()
        for item in response["items"]:
            comentario = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
            comentarios.append(comentario)
        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break
    return comentarios

# Función para escribir los comentarios en un archivo de texto
def escribir_comentarios_en_archivo(comentarios, nombre_archivo):
    with open(nombre_archivo, 'w', encoding='utf-8') as archivo:
        for comentario in comentarios:
            archivo.write(comentario + '\n')

# Obtener los IDs de todos los videos en la lista de reproducción
videos_ids = obtener_ids_videos()

# Iterar sobre los IDs de los videos y obtener los comentarios
for video_id in videos_ids:
    comentarios = obtener_comentarios(video_id)
    # Generar un nombre de archivo basado en el ID del video
    nombre_archivo = f"comentarios_video_{video_id}.txt"
    # Escribir los comentarios en el archivo correspondiente
    escribir_comentarios_en_archivo(comentarios, nombre_archivo)
    print(f"Comentarios del video {video_id} almacenados en {nombre_archivo}")




```
